```python
##Q2
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot a line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
##Q2
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Plot a box plot for age by pclass
sns.boxplot(x="pclass", y="age", data=titanic)

# Plot a box plot for fare by pclass
sns.boxplot(x="pclass", y="fare", data=titanic)
```

```python
##Q1
## Seaborn is a popular data visualization library in Python that provides a wide variety of plot types to visualize different kinds of data. Here are five plots that can be created using Seaborn, along with their uses:

## Scatter plot - A scatter plot is used to visualize the relationship between two variables. It displays points on a two-dimensional plane, where each point represents a pair of values for the two variables being plotted.

## Bar plot - A bar plot is used to display the distribution of a categorical variable, with each category represented as a bar whose height corresponds to the frequency or proportion of that category.

## Line plot - A line plot is used to visualize the trend or pattern in a continuous variable over time or across another continuous variable.

## Histogram - A histogram is used to visualize the distribution of a continuous variable by dividing the data into equal intervals (bins) and displaying the frequency or proportion of data points falling in each bin as a bar.

## Heatmap - A heatmap is used to visualize the relationship between two categorical variables or one categorical and one continuous variable. It displays the frequency or proportion of observations in a matrix of cells, with different colors representing different levels of frequency or proportion.
```

```python
##Q4
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")
# Plot a histogram of the "price" column with "cut" used as a hue
sns.histplot(data=diamonds, x="price", hue="cut")
```

```python
import seaborn as sns

# Load the iriss dataset
diamonds = sns.load_dataset("iris")

# Plot a pair plot "cut" used as a hue
sns.pairplot(data=diamonds, hue="species")
```

```python
import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix of the number of passengers by month and year
flights = flights.pivot("month", "year", "passengers")

# Plot a heatmap of the number of passengers by month and year
sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
```
